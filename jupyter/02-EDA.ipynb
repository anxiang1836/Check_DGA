# EDA分析

```python
import pickle
import pandas as pd
def load_data(dataPath):
    with open(dataPath,"rb") as f:
        df = pickle.load(f)
    return df

dga_360_Path = "data/dga_360"
alexa_2019_Path = "data/alexa_2019"

dga_360_df = load_data(dataPath=dga_360_Path)
alexa_2019_df = load_data(dataPath=alexa_2019_Path)
```

```python
import numpy as np
alexa_2019_df["url_len"] = alexa_2019_df["urls"].agg(lambda x:len(x))
dga_360_df["url_len"] = dga_360_df["urls"].agg(lambda x:len(x))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# URL长度分布

## alexa长度统计

```python
sns.distplot(alexa_2019_df["url_len"],kde=False)
```

```python
# 90、95、99分位数
np.percentile(alexa_2019_df["url_len"], [90,95,99])
```

## 360dga长度统计

```python
sns.distplot(dga_360_df["url_len"],kde = False)
```

```python
# 90、95、99分位数
np.percentile(dga_360_df["url_len"], [90,95,99])
```

## 360dga的family统计

```python
dga_360_family_count = pd.DataFrame(dga_360_df.groupby("family").\
                                    agg({'urls':'count','url_len':[np.min,np.max,np.median,np.mean]}))
dga_360_family_count = dga_360_family_count.reset_index()
dga_360_family_count.columns = ["family","count","min","max","median","avg"]
dga_360_family_count
```
